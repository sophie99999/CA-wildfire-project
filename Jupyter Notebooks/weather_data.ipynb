```python
import pandas as pd
import pymongo
```

```python
weather_df = pd.read_csv("./data_all.csv")
```

```python
weather_df.head()
```

```python
# df.drop('column_name', axis=1, inplace=True)
# weather_df.columns
weather_df.drop(['Unnamed: 0', 'CDSD', 'CLDD', 'DP01', 'DP10', 'DP1X',
       'DSND', 'DSNW', 'DT00', 'DT32', 'DX32', 'DX70', 'DX90', 'ELEVATION',
       'EMNT', 'EMSD', 'EMSN', 'EMXP', 'EMXT', 'HDSD', 'HTDD', 'SNOW', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
weather_df
```

```python
# DataFrame.set_index(self, keys, drop=True, append=False, inplace=False, verify_integrity=False)
weather_df.set_index('DATE', inplace=True)
```

```python
weather_df.head()
```

```python
yearly_avg = weather_df.groupby(['DATE', 'STATION'])
```

```python
yearly_avg_df = yearly_avg.mean()
```

```python
yearly_avg_df.fillna(value=0, axis=1, inplace=True)
```

```python
yearly_avg_df.head()
```

```python
# data_dict = data.to_dict('series') 
weather_dict = yearly_avg_df.to_dict()
```

```python
weather_dict['AWND'][2014, 'USW00094299']
```

```python
new_yearly_avg_df = yearly_avg_df
```

```python
new_yearly_avg_df.reset_index(level=1, inplace=True)
```

```python
new_yearly_avg_df.reset_index(level=0, inplace=True)
```

```python
weather_dict = new_yearly_avg_df.to_dict('records')
```

```python
weather_dict
```

```python
for station in weather_dict: 
    client = pymongo.MongoClient('localhost', 27017)
    db = client.wildfire
    yearly_weather_data = db.yearly_weather_data
    yearly_weather_data.insert_one(station)
```
