```python
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser("chrome", **executable_path)
fires_2019 = "https://www.fire.ca.gov/incidents/2019/"
url = fires_2019
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, "html.parser")

# year_collection = db.ca_wildfires_2019

wildfire2019 = []

for x in range(1,28):
    html = browser.html
    soup = BeautifulSoup(html,"html.parser")
    
    table = soup.find("div", class_="responsive-table")
    rows = table.find_all("div", class_="responsive-table-row")
    for row in rows:
        cells = row.find_all("div", class_="responsive-table-cell")
        temp_row_dict = {}
        for cell in cells:
            text_list = cell.text.split(":")
            if len(text_list) == 2: 
                temp_row_dict[text_list[0].strip()] = text_list[1].strip()
        
        wildfire2019.append(temp_row_dict)
        
    browser.click_link_by_text(str(x))
    
print(wildfire2019)
# for fire in wildfire2019:
#     dict_datbase = {}
#     for key, value in fire.items():
#         dict_database[key] = value
#     year_collection.insert_one(dict_database)
```
